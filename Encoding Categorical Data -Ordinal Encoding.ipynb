```python
import pandas as pd

# Provide the raw GitHub URL of the CSV file
url = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day26-ordinal-encoding/customer.csv"
df = pd.read_csv(url)
df=df.iloc[:,2:]
df.sample(5)
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
X_train



from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(X_train)
X_train=oe.transform(X_train)
X_test=oe.transform(X_test)
#similarly for label encoder
# it is used explicitly for the output column only 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder() # the values will be given randomly e.g. no->1 and yes->2
le.fit(y_train)
y_train=le.transform(y_train)
y_test=le.transform(y_test)
le.classes_ # to identify the values of the attributes
```
