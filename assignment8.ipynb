```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
iqr = 1.5 * (q3 - q1)
lb = q1 - iqr
ub = q3 + iqr
df = df.loc[(df['age'] >= lb) & (df['age'] <= ub)]
print(lb,ub)
```

```python
q1 = df['fare'].quantile(0.25)
q3 = df['fare'].quantile(0.75)
iqr = 1.5 * (q3 - q1)
lb = q1 - iqr
ub = q3 + iqr
df = df.loc[(df['fare'] >= lb) & (df['fare'] <= ub)]
print(lb,ub)
```

```python
df = df.drop(columns=['who'])
```

```python
df
```

```python
df.select_dtypes(include=['object']).keys()
```

```python
df
```

```python
df['age'] = df['age'].astype(int)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['alive'] = encoder.fit_transform(df['alive'])
df['class'] = encoder.fit_transform(df['class'])
df['embark_town'] = encoder.fit_transform(df['embark_town'])
df['embarked'] = encoder.fit_transform(df['embarked'])
df['sex'] = encoder.fit_transform(df['sex'])
df['alone'] = encoder.fit_transform(df['alone'])
df['adult_male'] = encoder.fit_transform(df['adult_male'])
df['deck'] = encoder.fit_transform(df['deck'])
```

```python
df
```

```python
x = df.drop(columns=['alive'])
y = df['alive']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier
DTmodel = DecisionTreeClassifier()
```

```python
DTmodel.fit(x_train,y_train)
```

```python
y_pred = DTmodel.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNNmodel = KNeighborsClassifier()
```

```python
KNNmodel.fit(x_train,y_train)
```

```python
y_pred = KNNmodel.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
RFmodel = RandomForestClassifier()
```

```python
RFmodel.fit(x_train,y_train)
```

```python
y_pred = RFmodel.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import BaggingClassifier
bagging_model = BaggingClassifier(estimator=RFmodel,n_estimators=10)
```

```python
bagging_model.fit(x_train,y_train)
```

```python
y_pred = bagging_model.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import StackingClassifier
estimators = [
    ('dt', DTmodel),
    ('knn', KNNmodel),
    ('rf', RFmodel)
]
stacking_model = StackingClassifier(estimators=estimators,final_estimator=RFmodel)
```

```python
stacking_model.fit(x_train,y_train)
```

```python
y_pred = stacking_model.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
boosting_model = GradientBoostingClassifier(n_estimators=100,learning_rate=0.2)
```

```python
boosting_model.fit(x_train,y_train)
```

```python
y_pred = boosting_model.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```
