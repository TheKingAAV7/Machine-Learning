```python
import os
import random
import shutil

class_folders = ['zero', '9', '8', '7', '6', '5', '4', '3', '2', '1']

data_dir=r"C:\Users\HP\Desktop\Output"
train_dir=os.path.join(data_dir,'train')
validation_dir=os.path.join(data_dir,'validation')
test_dir=os.path.join(data_dir,'test')

os.makedirs(train_dir,exist_ok=True)
os.makedirs(validation_dir,exist_ok=True)
os.makedirs(test_dir,exist_ok=True)

train_ratio=0.6
validation_ratio=0.2
test_ratio=0.3


for class_folder in class_folders:
    class_path =os.path.join(data_dir,class_folder)
    images=os.listdir(class_path)
    random.shuffle(images)
    
    # Determine the number of images for each split
    num_images=len(images)
    num_train=int(num_images * train_ratio)
    num_validation=int(num_images * validation_ratio)
    num_test=num_images - num_train - num_validation
    
    # Split the images
    train_images=images[:num_train]
    validation_images=images[num_train:num_train+num_validation]
    test_images=images[num_train + num_validation:]
    
    # Create subdirectories for each class in train, validation, and test directories
    train_class_dir = os.path.join(train_dir,class_folder)
    validation_class_dir = os.path.join(validation_dir,class_folder)
    test_class_dir = os.path.join(test_dir,class_folder)
    
    os.makedirs(train_class_dir, exist_ok=True)
    os.makedirs(validation_class_dir, exist_ok=True)
    os.makedirs(test_class_dir, exist_ok=True)
    
    # Move the images to the appropriate subdirectories
    
    for image in train_images:
        source=os.path.join(class_path,image)
        destination=os.path.join(train_class_dir,image)
        shutil.copy(source,destination)
        
    for image in validation_images:
        source=os.path.join(class_path,image)
        destination=os.path.join(validation_class_dir,image)
        shutil.copy(source,destination)
        
    for image in test_images:
        source=os.path.join(class_path,image)
        destination=os.path.join(test_class_dir,image)
        shutil.copy(source,destination)
        
```

```python
import os
import numpy as np

class_folders = ['zero', '1', '2', '3', '4', '5', '6', '7', '8', '9']

# Initialize empty lists to store labels and images
labels = []
images = []

# Map class folder names to numeric labels
class_to_label = {class_name: class_label for class_label, class_name in enumerate(class_folders)}

data_dir = r"C:\Users\HP\Desktop\Output"

# Iterate through each class folder and its corresponding label
for class_folder in class_folders:
    class_path = os.path.join(data_dir, class_folder)
    class_label = class_to_label[class_folder]
    
    # List all image files in the class folder
    images_in_class = os.listdir(class_path)
    
    # Append the class label to the 'labels' list for each image in this class
    labels.extend([class_label] * len(images_in_class))
    
    # Append the image data to the 'images' list
    for image_filename in images_in_class:
        image_path = os.path.join(class_path, image_filename)
        image = cv2.imread(image_path)
        # You may want to add image preprocessing here (e.g., resizing)
        images.append(image)

# Convert the 'labels' and 'images' lists to NumPy arrays
labels_array = np.array(labels)
images_array = np.array(images)

# You can save these label and image arrays for later use if needed
np.save('labels.npy', labels_array)
np.save('images.npy', images_array)


```
